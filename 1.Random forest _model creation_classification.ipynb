```python
import pandas as pd
```

```python
dataset=pd.read_csv("Social_Network_Ads.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
independent=dataset[[ 'Age', 'EstimatedSalary',  'Gender_Male']]
independent
```

```python
dependent=dataset[['Purchased']]
dependent
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
x_train
```

```python

from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
param_grid={'n_estimators':[100,1000,200],'criterion':['entropy']}
grid=GridSearchCV(RandomForestClassifier(),param_grid,refit=True,verbose=3,n_jobs=-1)

grid.fit(x_train,y_train)
```

```python
grid_prediction=grid.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,grid_prediction)
```

```python
print("The report of cm : \n", cm)
```

```python
from sklearn.metrics import classification_report
clf_report=classification_report(y_test,grid_prediction)
```

```python
print("The report of Classification report : \n",clf_report)
```
